```python
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np 
import matplotlib as mp
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf 

sess = tf.Session()
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
imageWidth = imageHeight = 28
num1 = 3   
imageToUse1 = mnist.test.images[num1]
reshaped1 = np.reshape(imageToUse1,[imageWidth,imageHeight])

num2 = 4   
imageToUse2 = mnist.test.images[num2]
reshaped2 = np.reshape(imageToUse2,[imageWidth,imageHeight])

num3 = 5   
imageToUse3 = mnist.test.images[num3]
reshaped3 = np.reshape(imageToUse3,[imageWidth,imageHeight])

num4 = 6   
imageToUse4 = mnist.test.images[num4]
reshaped4 = np.reshape(imageToUse4,[imageWidth,imageHeight])


plt.subplot(221)
plt.imshow(reshaped1, cmap=plt.get_cmap('gray'))
plt.subplot(222)
plt.imshow(reshaped2, cmap=plt.get_cmap('gray'))
plt.subplot(223)
plt.imshow(reshaped3, cmap=plt.get_cmap('gray'))
plt.subplot(224)
plt.imshow(reshaped4, cmap=plt.get_cmap('gray'))
# show the plot
plt.show()
```
