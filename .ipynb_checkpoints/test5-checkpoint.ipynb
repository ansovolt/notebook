```python
%matplotlib inline
from pyimagesearch.transform import four_point_transform
from pyimagesearch import imutils
from skimage.filters import threshold_otsu,threshold_adaptive
import numpy as np
import argparse
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from skimage import data
```

```python
img = cv2.imread('document-scanner/images/1-original-receipt-view.png')
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
titles = ['Original Image', 'Global Thresholding (v = 127)','Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]
for i in xrange(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python

#image = cv2.imread('document-scanner/images/1-original-receipt-view.png')
#image = data.page()
im = cv2.imread('document-scanner/images/1-original-receipt-view.png')
image = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

global_thresh = threshold_otsu(image)
binary_global = image > global_thresh

block_size = 40
binary_adaptive = threshold_adaptive(image, block_size, offset=10)

fig, axes = plt.subplots(nrows=3, figsize=(7, 8))
ax0, ax1, ax2 = axes
plt.gray()

ax0.imshow(image)
ax0.set_title('Image')

ax1.imshow(binary_global)
ax1.set_title('Global thresholding')

ax2.imshow(binary_adaptive)
ax2.set_title('Adaptive thresholding')

for ax in axes:
    ax.axis('off')

plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_GBR2GRAY)
gray = cv2.GaussianBlur(gray, (5, 5), 0)
edged = cv2.Canny(gray, 75, 200)
```

```python
plt.imshow(image)
plt.show()
plt.imshow(edged, cmap='gray')
plt.show()
```

```python
_, cnts, hierarchy = cv2.findContours(edged.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
cnts = sorted(cnts, key = cv2.contourArea, reverse = True)[:10]
screenCnt = None
i = 0
maxp = 0
selp = 0
selc = None
for c in cnts:
	#print ("countur %d" % i)
	#i = i + 1
	# approximate the contour
	peri = cv2.arcLength(c, True)
	approx = cv2.approxPolyDP(c, 0.02 * peri, True)
	#print ("peri is %s" % peri)
#	if selp < peri:
#		selp = peri
#		selc = approx
	#print ("approx is %s" % approx) 
	# if our approximated contour has four points, then we
	# can assume that we have found our screen
	if len(approx) == 4:
		#print ("found countur")
		screenCnt = approx
		#print(type(screenCnt))
		break
#print("selp %s" % selp)
```
